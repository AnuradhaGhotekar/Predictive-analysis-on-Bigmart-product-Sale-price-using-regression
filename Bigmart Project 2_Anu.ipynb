# Feature selection

```python
Y = A[["Item_Outlet_Sales"]]
A = A.drop(columns=["Item_Identifier","Outlet_Identifier"],axis=1)
```

# Missing data treatment and Cat, Con seperation

```python
from PM6 import seperate,ANOVA,replacer
replacer(A)
q = list(A.columns)
q.remove("Item_Outlet_Sales")

con,cat =seperate(A)
```

```python
cat
```

```python
con
```

```python
imp_cat = []
q=list(cat)
for i in q :    
        print("--------------------")
        print("Item_Outlet_Sales vs",i)
        w = ANOVA(A,i,"Item_Outlet_Sales")
        print(w)
        if(w < 0.05):
            imp_cat.append(i)    
```

```python
imp_con = []
Q = A[con].corr()["Item_Outlet_Sales"].sort_values()
imp_con = list(Q[(Q < 1)&(Q > 0.3)].index)
print(Q)
```

```python
imp_cat
```

```python
imp_con
```

```python
A = A[imp_cat].join(A[imp_con])
```

```python
A.head()
```

```python
A.shape
```

```python
from PM6 import preprocessing
X =preprocessing(A)
X.head()
```

```python
X
```

```python
X.skew().sort_values
```

```python
S = X.skew()
skew = list(S[S>2].index)
skew
```

```python
import numpy as np
from numpy import log
for j in skew:
    W = []
    for i in X[j]:
        if(i != 0):
            W.append(np.log(i))
        else:
            W.append(i)
    X[j] = W
```

```python
Q =['Item_MRP', 'Item_Type_Baking Goods', 'Item_Type_Breads',
        'Item_Type_Breakfast', 'Item_Type_Canned', 'Item_Type_Dairy',
        'Item_Type_Frozen Foods', 'Item_Type_Fruits and Vegetables',
        'Item_Type_Hard Drinks', 'Item_Type_Health and Hygiene',
        'Item_Type_Household', 'Item_Type_Meat', 'Item_Type_Others',
        'Item_Type_Seafood', 'Item_Type_Snack Foods', 'Item_Type_Soft Drinks',
        'Item_Type_Starchy Foods', 'Outlet_Size_High', 'Outlet_Size_Medium',
        'Outlet_Size_Small', 'Outlet_Location_Type_Tier 1',
        'Outlet_Location_Type_Tier 2', 'Outlet_Location_Type_Tier 3',
        'Outlet_Type_Grocery Store', 'Outlet_Type_Supermarket Type1',
        'Outlet_Type_Supermarket Type2', 'Outlet_Type_Supermarket Type3']
```

```python
Q[0:2]
```

```python
for i in Q:
    j = Q[0:Q.index(i)+1]
    Xnew = X[j]
    print("Selected columns:",Xnew.columns)
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size=0.2,random_state=21)
    from sklearn.linear_model import LinearRegression
    lm=LinearRegression()
    model = lm.fit(xtrain,ytrain)
    pred = model.predict(xtest)
    from sklearn.metrics import mean_squared_error
    
    mse1 = round(mean_squared_error(ytest,pred),4)
    print("\n Mean squared error is:",mse1)
    
    
```

# As last value of mean squared error is minumum out of all, Model with all coulmns is  the best model  

```python
Xnew = X
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(Xnew,Y,test_size=0.2,random_state=21)
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
model = lm.fit(xtrain,ytrain)
pred = model.predict(xtest)
from sklearn.metrics import mean_squared_error
    
mse1 = round(mean_squared_error(ytest,pred),4)
print("\n Mean squared error is:",mse1)
```
